```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
df = pd.read_csv('loop_track_waypoints.csv') # Load the CSV file into a DataFrame
index = df['Index'].values
x = df['X'].values
y = df['Y'].values
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Track Waypoints')
```

```python
f_x = sp.interpolate.interp1d(index ,x, kind='cubic') # Used cubic spline for x and index as x is looping
f_y = sp.interpolate.interp1d(index ,y, kind='cubic') # Used cubic spline for y and index as y is looping
t_new = np.linspace(index.min(), index.max(), num=1000) # Generate new index values for plotting so that it looks smooth
x_new = f_x(t_new)
y_new = f_y(t_new)
plt.plot(x_new, y_new, label='Cubic Spline Interpolation', color='orange', linewidth=2)
plt.scatter(x, y, label='Original Waypoints', color='blue')
plt.legend()
plt.show()
```
